```python
#import pandas and read a vine_table csv
import pandas as pd
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
#create a new df of filtered data
new_vine_df = vine_df.loc[vine_df['total_votes']>= 20]
new_vine_df.head()
```

```python
#create a new column with a percentage to then filter by for a new df
new_vine_df['percentage_votes'] = new_vine_df['helpful_votes']/new_vine_df['total_votes']
nn_vine = new_vine_df.loc[new_vine_df['percentage_votes']>= 0.5]
nn_vine.head()
```

```python
#drop the created percentage_vote column from the new df
nn_vine_df = nn_vine.drop(['percentage_votes'], axis=1)
nn_vine_df.head()
```

```python
#paid reviews
vine = nn_vine_df.loc[nn_vine_df['vine'] == 'Y']
vine.head()
```

```python
#not paid reviews
vine_n = nn_vine_df.loc[nn_vine_df['vine'] == 'N']
vine_n.head()
```

```python
#determine total review number,the number of 5-star reviews, and the percentage of 5-star review for paid 
vine.info()
```

```python
#determine total review number,the number of 5-star reviews, and the percentage of 5-star review for paid 
vine_star = vine.groupby(['star_rating'])['star_rating'].count()
print(vine_star)
#percentage of the 5-star reviews is 65/170*100 = 38.23%
```

```python
#determine total review number,the number of 5-star reviews, and the percentage of 5-star review for not paid 
vine_n.info()
```

```python
#determine total review number,the number of 5-star reviews, and the percentage of 5-star review for not paid 
vine_n_star = vine_n.groupby(['star_rating'])['star_rating'].count()
print(vine_n_star)
#percentage of the 5-star reviews is 20612/37840*100 = 54.47%
```
